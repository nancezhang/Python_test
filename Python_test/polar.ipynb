```python
#这是极坐标图形
%matplotlib inline
import numpy as np
from matplotlib.pyplot import figure, rc

rc('grid', color='g', lw=1, ls='-')
rc('xtick', labelsize=15)
rc('ytick', labelsize=15)

fig = figure(figsize=(15, 15))
ax = fig.add_axes([0.2, 0.2, 0.8, 0.8],projection='polar', fc='w')

r = np.arange(0, 4, 0.01)
c= 2*np.pi*r
ax.plot(c, r, color='r', lw=4, label='real line')
ax.plot(0.5*c, r, color='blue', ls='--', lw=4, label='broken line')
ax.legend()
```
